Tidak ditemukan adanya kolom dengan nama kolom dan isinya yang kurang sesuai.

```python
# pengelompokan kolom berdasarkan jenisnya
cats = ['CustomerID', 'PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus']
nums = ['Churn', 'Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress', 'Complain', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount']
nums1 = ['Churn', 'Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore']
nums2 = ['NumberOfAddress', 'Complain', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount']
```

```python
df[nums].describe()
```

Tidak ditemukan adanya kolom yang memiliki nilai summary agak aneh, semua kolom terlihat normal.

```python
df[cats].describe()
```

Tidak ada duplikat dalam kolom 'CustomerID', semua kolom terlihat normal.

```python
for col in cats:
    print(f'''Value count kolom {col}:''')
    print(df[col].value_counts()) 
    print()
```

```python
for col in cats:
    print(f'''Percentage value count kolom  {col}:''')
    print(df[col].value_counts()/sum(df[col].value_counts())*100) 
    print()
```

# 2. Univariate Analysis

Mencari outlier dan melihat distribusi data

## Boxplot

```python
for i in range(0, len(nums1)):
    plt.subplot(1, len(nums1), i+1)
    sns.boxplot(y=df[nums1[i]], color='gray', orient='v')
    plt.tight_layout()
```

```python
for i in range(0, len(nums2)):
    plt.subplot(1, len(nums2), i+1)
    sns.boxplot(y=df[nums2[i]], color='gray', orient='v')
    plt.tight_layout()
```

## Distplot

```python
plt.figure(figsize=(12, 5))
for i in range(0, len(nums1)):
    plt.subplot(2, len(nums1), i+1)
    sns.distplot(df[nums1[i]], color='gray')
    plt.tight_layout()
```

```python
plt.figure(figsize=(12, 5))
for i in range(0, len(nums2)):
    plt.subplot(2, len(nums2), i+1)
    sns.distplot(df[nums2[i]], color='gray')
    plt.tight_layout()
```

## Violinplot

```python
for i in range(0, len(nums1)):
    plt.subplot(1, len(nums1), i+1)
    sns.violinplot(y=df[nums1[i]], color='gray', orient='v')
    plt.tight_layout()
```

```python
for i in range(0, len(nums2)):
    plt.subplot(1, len(nums2), i+1)
    sns.violinplot(y=df[nums2[i]], color='gray', orient='v')
    plt.tight_layout()
```

## Countplot

Melihat data categorical

```python
for i in range(0, len(cats[2:])):
    plt.subplot(2, 3, i+1)
    sns.countplot(df[cats[2:][i]], color='gray', orient='h')
    plt.tight_layout()
```
